### Read parameters from config file

```python
import configparser

# Initialize the ConfigParser object
config = configparser.ConfigParser()
# Read the INI file
config.read('track_breeding_terns_runner.ini')
# A date to track breeding terns on
date = config.get('General', 'date').split(',')
# Path to the input single-scan tracking results directory
one_scan_result_dir = config.get('General', 'one_scan_result_dir')
# Path to the output directory for multi-scan tracking results
mult_scans_result_dir = config.get('General', 'mult_scans_result_dir')
# Labels distribution of flags path
video_converter_dir = config.get('General', 'video_converter_dir')
classification_dir = config.get('General', 'classification_dir')
```

```python
print ("date:", date)
print("one_scan_result_dir:", one_scan_result_dir)
print("mult_scans_result_dir:", mult_scans_result_dir)
print("video_converter_dir:", video_converter_dir)
print ("classification_dir:", classification_dir)
```

```python
from datetime import datetime, timedelta
import glob

# parse INI date
track_date = datetime.strptime(date[0], "%Y_%m_%d").date()

# generate list of acceptable dates: today + 4 previous days
accepted_dates = [(track_date - timedelta(days=i)).strftime("%Y_%m_%d") for i in range(0, 5)]
print("Accepted dates:", accepted_dates)

def isFileFromAcceptedDates(path):
    return any(d in path for d in accepted_dates)

# List all directories under one_scan_result_dir
all_dirs = glob.glob(os.path.join(one_scan_result_dir, '*', '*'))

def get_dir_path_with_parent(directory_path):
  # Get the parent directory and the base name of the directory
  parent_dir, dir_name = os.path.split(os.path.normpath(directory_path))

  # Get the parent directory name and concatenate it with the directory name
  return os.path.join("/", os.path.basename(parent_dir), dir_name)[1:]

two_level_dirs = [
    get_dir_path_with_parent(path)
    for path in all_dirs
    if os.path.isdir(path) and isFileFromAcceptedDates(path)
]

print("My tour list looks like this:")
two_level_dirs
```

#### Amaglate tours which are from the same date

```python
# This function extract the camera number (181/191) from the tour string sub path
def get_camera_number(tour):
    return tour.split('.')[0].split('cam')[-1]

# This function Extracts the date from the tour string.
def get_tour_date(tour):
    return tour.split('_', 1)[1][:10]

def add_day_before_suffix(tour, today):
    """Mark tours from previous days with _day_before suffix."""
    if get_tour_date(tour) != today and not tour.endswith("_day_before"):
        return tour + "_day_before"
    return tour
```

```python
from collections import defaultdict

tours_by_cam = defaultdict(list)

for tour in two_level_dirs:
    cam = get_camera_number(tour)
    tours_by_cam[cam].append(tour)

print("Tours grouped by camera:")
for cam, tours in tours_by_cam.items():
    print(f"Camera {cam}: {len(tours)} tours")
```

```python
import subprocess

for cam, tours_list in tours_by_cam.items():
    print(f"\n🚀 Tracking camera {cam} with {len(tours_list)} tours")

    # Add _day_before to previous-day tours
    marked_tours = [add_day_before_suffix(t, date[0]) for t in tours_list]

    #print("Tours passed to tracker:")
    #for t in marked_tours:
    #    print("  ", t)

    cmd = ["python", "./track_breeding_terns_runner.py"] + marked_tours
    cmd_str = " ".join(cmd)

    result = subprocess.run(cmd_str, shell=True, capture_output=True, text=True)
    print(result.stdout if result.returncode == 0 else result.stderr)
```
