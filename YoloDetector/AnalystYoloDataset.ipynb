```python
import os
from collections import defaultdict

map_label = {
    '0': 'Chick',
    '1': 'Common-Fly',
    '2': 'Common-Sit',
    '3': 'Common-Stand',
    '4': 'Little-Fly',
    '5': 'Little-Sit',
    '6': 'Little-Stand',
    '7': 'Other',
}

# Base directory
base_dir = '/content/drive/MyDrive/tern_project/Eyal/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/'

# Subdirectories to look into
subdirs = ['train/labels', 'valid/labels', 'test/labels']

# Dictionary to count annotations
annotation_counts = defaultdict(int)

# Iterate over each subdirectory
for subdir in subdirs:
    dir_path = os.path.join(base_dir, subdir)
    if not os.path.exists(dir_path):
        print(f"Directory {dir_path} does not exist.")
        continue

    # Iterate over all text files in the directory
    for file_name in os.listdir(dir_path):
        if file_name.endswith('.txt'):
            file_path = os.path.join(dir_path, file_name)

            # Read the file and count annotations
            with open(file_path, 'r') as file:
                for line in file:
                    parts = line.strip().split()
                    if parts:  # Ensure the line is not empty
                        category = parts[0]  # First column is the category
                        annotation_counts[map_label[category]] += 1

```

```python
annotation_sum = sum(annotation_counts.values())
print(f"Annotation Number: {annotation_sum}")
print()

# Print the results
print("Annotation Counts:")
for category, count in annotation_counts.items():
    print(f"{category}: {count}")
```
